```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./giraffe.png")
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_grey , cmap="grey")
```

```python
kernel_s = 16

avg_blur = cv2.blur(image_grey,(kernel_s,kernel_s))

plt.imshow(avg_blur,cmap="grey")
```

```python
kernel_s = 51
image_gaussian = cv2.GaussianBlur(image_grey,(kernel_s,kernel_s),0)

plt.imshow(image_gaussian, cmap="grey")
```

```python
image_median = cv2.medianBlur(image_grey,23)

plt.imshow(image_median,cmap="grey")
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.title("original")
plt.imshow(image_grey , cmap="grey")
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Average Blur")
plt.imshow(avg_blur , cmap="grey")
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Gaussian Blur")
plt.imshow(image_gaussian , cmap="grey")
plt.axis("off")

plt.subplot(2,2,4)
plt.title("Median Blur")
plt.imshow(image_median , cmap="grey")
plt.axis("off")
```
