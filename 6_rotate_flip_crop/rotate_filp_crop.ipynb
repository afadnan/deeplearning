```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./wolf.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
#print(image)
#plt.imshow(image)
#print(image.shape)
```

```python
plt.imshow(image_rgb)
```

# Filp

```python
# 1 --> horizontally
# 0 --> Vertically
# -1 --> Both axes

flip_horizontal = cv2.flip(image,1)
flip_vertical = cv2.flip(image,0)
flip_both = cv2.flip(image,-1)

titles = ["Original image","Horizontal Flip","Vertical Flip","Flip Both Axes"]
images = [image_rgb,
          cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_vertical,cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB)
          ]

plt.figure(figsize=(12,8))
for i in range(4):
        plt.subplot(2,2,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis("Off")
plt.tight_layout()
plt.show()
```

```python
rotate_90_clockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb,cv2.ROTATE_180)

titles = ["Original image","90 degree clockwise","90 degree counterclockwise","180 degree"]
images = [image_rgb,
          rotate_90_clockwise,
          rotate_90_counterclockwise,
          rotate_180
          ]
plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("Off")
plt.tight_layout()
plt.show()
```
